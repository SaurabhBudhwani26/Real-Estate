```python
housing_data = strat_train_X.drop("MEDV", axis = 1)
housing_label = strat_train_X["MEDV"].copy()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy="median")
imputer.fit(housing_data)
```

```python
imputer.statistics_
```

```python
X = imputer.transform(housing_data)
housing_hg=pd.DataFrame(X, columns=housing_data.columns)
```

# Creating a pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy = "median")),
    ('std scaler', StandardScaler())
])
```

```python
housing_dt = my_pipeline.fit_transform(housing_data)
```

```python
housing_dt
```

# Selecting the desired model

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(housing_dt,housing_label)
```

# Evaluating the model

```python
from sklearn.metrics import mean_squared_error
housing_predictions = model.predict(housing_dt)
mse = mean_squared_error(housing_label, housing_predictions)
mse
```

# Cross-Validation

```python
from sklearn.model_selection import cross_val_score
scores= cross_val_score(model,housing_dt,housing_label, scoring="neg_mean_squared_error",cv=10)
rmse_score = np.sqrt(-scores)
```

```python
rmse_score
```

```python
def print_scores(scores):
    print("Scores:",scores)
    print("Mean:",scores.mean())
    print("Standard deviation:",scores.std())
```

```python
print_scores(rmse_score)
```
